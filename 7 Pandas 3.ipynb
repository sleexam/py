```python
data1=pd.DataFrame({'std_id':['S1','S2','S3','S4'],
                        'name':['Bumrah','Jadeja','Rohith','Rahul'],
                        'marks':[98,97,88,95]})
data2=pd.DataFrame({'std_id':['S3','S4','S5','S6'],
                        'name':['Jaiswal','Kohli','Pant','Ashwin'],
                        'marks':[97,95,91,92]})
```

```python
merge1=pd.merge(data1,data2,on='std_id',how='inner')
merge1
```

```python
merge2=pd.merge(data1,data2,on='std_id',how='outer')
merge2
```

```python
merge3=pd.merge(data1,data2,on='std_id',how='left')
merge3
```

```python
merge4=pd.merge(data1,data2,on='std_id',how='right')
merge4
```

```python
frame1=pd.DataFrame({'id':['Ball','Pencil','pen','mug','ashtray'],
                     'color':['red','blue','blue','black','green'],
                     'brand':['omg','abc','abc','pod','pod']})
frame2=pd.DataFrame({'id':['Pencil','Pencil','Ball','pen'],
                     'brand':['abc','pod','abc','pod']})
merge=pd.merge(frame1,frame2)
merge
```

```python
df1=pd.merge(frame1,frame2,on='id',how='inner')
df1
```

```python
df2=pd.merge(frame1,frame2,on='id',how='outer')
df2
```

```python
df3=pd.merge(frame1,frame2,on=['id','brand'],how='inner')
df3
```

```python
df4=pd.merge(frame1,frame2,on=['id','brand'],how='outer')
df4
```

```python
dp=pd.DataFrame({'name':['Ram','Varun','sudhu','Uthappa','Suhas'],
                 'score':[90,60,70,np.NaN,80],
                 'qualify':['Yes','yes','yes','No','Yes']},
                index=['a','b','c','d','e'])
dp
```
