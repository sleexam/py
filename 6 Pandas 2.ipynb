```python
data['Total']=data['Python']+data['Machine Learning']+data['Java']+data['IOT']+data['SAN']
data
```

```python
plt.hist([data['Python'],data['Machine Learning'],data['Java'],data['IOT'],data['SAN']],color=['red','green','yellow','pink','brown'])    
plt.xlabel('Score')
plt.ylabel('No of students')
plt.legend(['Python','Machine Learning','Java','IOT','SAN'])
plt.show()
```

```python
y=['Python','Machine Learning','Java','IOT','SAN']
width=[data['Python'].max(),data['Machine Learning'].max(),data['Java'].max(),data['IOT'].max(),data['SAN'].max()]
plt.figure(figsize=(12,2))
plt.barh(y=y,width=width)
plt.title('Average Score')
plt.xlabel('Average Score')
plt.ylabel('Subjects')
for i,v in enumerate(width):
    plt.text(v,i," "+str(round(v)), color='green', va='center', fontweight='bold')
plt.show()
```

```python
Gen=data.groupby('Gender')
y=Gen['Gender'].count().keys()
width=Gen['Gender'].count()
plt.figure(figsize=(12, 2))
plt.barh(y=y, width=width)
plt.title('Number of Males and Females')
plt.xlabel('Count')
plt.ylabel('Gender')
for i,v in enumerate(width):
    plt.text(v, i, " " +str(round(v)), color='green', va='center', fontweight='bold')
plt.show()
```
